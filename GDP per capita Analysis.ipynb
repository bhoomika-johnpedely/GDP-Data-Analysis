## GDP Per Capita Analysis

In this project, I have analyzed GDP Per Capita (Price and Inflation adjusted) for countries all over the world. It is given in terms of PPP (Purchasing Power Parity) across countries over many years.  

### Data Collection

The data used in this analysis is from Gapminder. https://www.gapminder.org/data/documentation/gd001/. The dataset contains GDP for all the countries from 1800 and also includes projections (by gapminder) until 2100.

### Data Exploration and Transformation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data
url = "https://github.com/bhoomika-johnpedely/GDP-Data-Analysis/blob/main/gdp_pcap.csv?raw=True"
data = pd.read_csv(url)
```

```python
# View the data
df = data
df
```

```python
# Removing columns I am not interested in
columns_to_remove = [str(year) for year in range(1800, 1923)] + [str(year) for year in range(2023, 2101)]

df = df.drop(columns=columns_to_remove)
```

```python
#Viewing the new df
df
```

```python
df.dtypes
```

```python
#converting all 'k' values to proper numeric terms
def k_to_thousands(value):
    if 'k' in value:
        return float(value.replace('k', '')) * 1000
    return int(value)

# applying the function to all numeric columns
for i in range(1, len(df.columns)):
    df.iloc[:, i] = df.iloc[:, i].apply(k_to_thousands)

df
```

```python
# all values are float, converting them to int
for col in df.columns[1:]:
    df[col] = df[col].astype(int)
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Checking for null
df.isnull().sum()
```

```python
#Checking for duplicates
df.duplicated().sum()
```

```python
#Getting the countries in order
sorted_df = df.sort_values(by='country')
sorted_df.head()
```

```python
#Setting the countries as index so its easier to look up a specific country
new_df = sorted_df.set_index('country')
new_df.head()
```

##### Comment 1

As part of cleaning the data, I have only kept columns for the last 100 years, converted any string values to proper numeric values and checked for null and duplicates. I have also sorted the dataframe and set countries as the index for rows.

```python
#Top 5 countries the last 3 years
last_3_years = ['2020', '2021', '2022']

for year in last_3_years:
    top5_gdp = century_df[year].nlargest(n=5)
    print(year, top5_gdp)
```

```python
# Comparative GDP growth for top 5 over 100 years
top_countries = ['Monaco','Luxembourg','Singapore','Ireland','Qatar']

for country in top_countries:

    gdp = century_df.loc[country]

    plt.plot(gdp)
    plt.xticks([])
    plt.legend(top_countries)

plt.ylabel('GDP')
plt.title('GDP Over The Last 100 Years: Top 5 Countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Fastest growing gdps over the last 10 years
```

```python
# New df with GDP for the last 10 years
columns_to_remove = [str(year) for year in range(1923, 2013)]

decade_df = century_df.drop(columns=columns_to_remove) 
decade_df
```

```python
# percentage change in every year for all the countries
growth_df = decade_df.pct_change(axis='columns', periods=1)
growth_df
```

```python
#Creating a column with the average percentage change of 10 years
growth_df['Average Growth'] = growth_df.mean(axis=1)
```

```python
growth_df
```

```python
#Top 10 countries by average growth
fastest_growing = growth_df['Average Growth'].nlargest(10)
fastest_growing
```

```python
#Fastest growing countries by GDP
palette = sns.color_palette("Set2")

fastest_growing.plot(kind='bar', color = palette)

plt.title('Fastest Growing Countries by GDP')
plt.xlabel('Country')
plt.ylabel('Average Percentage Growth in GDP')

plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

##### Comment 6

<p style="font-family: Times New Roman;font size: 14px;">
Guyana is, by far, the fastest growing country by average GDP growth. Ireland and China are next.</p>

```python
#USA GDP and growth rate
usa_gdp = century_df.loc['USA', '2022']
print(usa_gdp)
usa_growth = growth_df.loc['USA','Average Growth']
print(usa_growth * 100)
```

##### Comment 7

<p style="font-family: Times New Roman;font size: 14px;">
In 2022, USA's GDP is 64,599 and average growth rate for the last 10 years is 1.55%</p>

```python
#Let's see how the 3 fastest growing countries compare to the USA. 
US_vs_fastest = ['USA','Guyana','Ireland','China']

for country in US_vs_fastest:

    gdp = century_df.loc[country]

    plt.plot(gdp)
    plt.xticks([])
    plt.legend(US_vs_fastest)

plt.ylabel('GDP')
plt.title('USA and the fastest growing countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```
